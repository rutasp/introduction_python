```python
#Cleaning data
#drop NA rows where  values are NaN
df.dropna(axis='rows', inplace=True)
df
```

```python
#cleaning data
#delete not needed columns
columns = ['area',
'area_type',]
df = df.drop(columns, axis=1)
df
```

```python
#Add new columns with calculations from current
# df['Total_cases'] = df['vaccinated_cases'] + df['unvaccinated_cases']
# df['Vaccinated_%_in_hospitals'] = df['vaccinated_hosp'] / df ['vaccinated_cases'] 
# df['Total population'] = df['population_unvaccinated'] + df['population_vaccinated']
df["new0"] = df.vaccinated_deaths.where(df.vaccinated_deaths >= 1, "No death of vaccinated")
df['new1']=np.subtract(df['unvaccinated_hosp'],df['unvaccinated_deaths'])
df['new2']=np.random.randint(333)
df['new3']=np.where(df['unvaccinated_cases']>df['vaccinated_cases'],True, False)
df['new4']=np.round(df['unvaccinated_deaths_per_100k'],2)
df['new5']=np.add(df['vaccinated_cases'],df['unvaccinated_cases'])
df
```

```python
#filtering
# df = df[ ( df ['vaccinated_cases'] > 100 ) & ( df ['new3'] == True ) | ( df ['new5'] <10000 ) ]
df = df[ ( df ['unvaccinated_cases'] > 1 ) & ( df ['unvaccinated_hosp_per_100k'] <10 ) ]
df
```

```python
#adding this to be able to create pivot
df['Total_Population'] = df['population_unvaccinated'] + df['population_vaccinated']
df['More_cases']=np.where(df['unvaccinated_cases']>df['vaccinated_cases'],'Unvaccinated', 'Vaccinated')
df['Random_filter']=np.where(df['unvaccinated_cases_per_100k']>400, True, False)
df['even_odd']= np.where(df['Total_Population'] % 2 == 0, "More", "Less")
df['pop_vacc_even_odd']=np.where(df['population_vaccinated'] % 2 == 0, "Pfizer", "AstraZeneca")
df
```

```python
table = pd.pivot_table(
    df,  # data table
    values= "new5", # values of total cases
    index= ['More_cases', "pop_vacc_even_odd"],
    columns='even_odd',
    aggfunc=np.sum  #aggregation  table
 )
```
